#My First Pytorch Project: 1-Layer Linear Regression Model

## Introduction

I'm learning Pytorch and basic Linear Regression to understand details before the deep learning magic.

### Imports

```python
import torch
import matplotlib.pyplot as plt
```

### Creating Tensors

```python
# 2D Tensors
x = torch.tensor([[1.0], [2.0], [3.0], [4.0]])
y_true = torch.tensor([[2.0], [4.0], [6.0], [8.0]])
```

```python
# Creating a model
model = torch.nn.Linear(1, 1)

# Loss Function
loss_fn = torch.nn.MSELoss()

# Optimization = Algebraic Correction
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# Training Loop
for epoch in range(500):
    y_pred = model(x)
    loss = loss_fn(y_pred, y_true)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

# Weight and Bias
w = model.weight.item()
b = model.bias.item()

# Extract Learned Parameters
print(w, b)
```

### Visualize with Matplotlib

```python
# Converting Tensor to Numpy
x_np = x.numpy()
y_np = y_true.numpy()

# Generate Regression Line
x_line = torch.linspace(x.min(), x.max(), 100).view(-1, 1)
y_line = model(x_line).detach()

# Scatter Plot
plt.scatter(x_np, y_np, label='Data')
plt.plot(x_line, y_line, color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
